```python
class Chocolate:
  def __init__(self, price: float, additive: str = ""):
    if price <= 0:
      raise ValueError("Цена батончика не может быть равна нулю")
    self.price = price
    self.additive = additive

  def show_my_chocolate(self):
    if self.additive:
      print(f"Шоколадный батончик, добавка: {self.additive}")
    else:
      print("Обычный шоколадный батончик")

  def __str__(self):
    return f"Шоколадный батончик, {self.price} рублей"

try:
  Chocolate_dark = Chocolate(75.5, "Фундук")
  Chocolate_dark.show_my_chocolate()
  print(Chocolate_dark)

  Chocolate_white = Chocolate(110.0)
  Chocolate_white.show_my_chocolate()
  print(Chocolate_white)

  Chocolate_milk = Chocolate(0)
  Chocolate_milk.show_my_chocolate()
  print(Chocolate_milk)

except ValueError as error:
    print(error)

```
